```python
import tensorflow as tf
import numpy as np
```

```python
zero_dim = tf.constant(4)
print(zero_dim)

one_dim = tf.constant([2, 0, -3])
print(one_dim)

two_dim = tf.constant([
    [56, 67, -3],
    [4, 88, -100]
])
print(two_dim)

three_dim = tf.constant([
    [[56, 67, -3],
    [4, 88, -100]],

    [[56, 67, -3],
    [4, 88, -100]],

    [[56, 67, -3],
    [4, 88, -100]],

    [[56, 67, -3],
    [4, 88, -100]]
])
print(three_dim)

four_dim = tf.constant([
    [
      [[56, 67, -3],
      [4, 88, -100]],

      [[56, 67, -3],
      [4, 88, -100]],

      [[56, 67, -3],
      [4, 88, -100]],

      [[56, 67, -3],
      [4, 88, -100]]
    ],

    [
      [[56, 67, -3],
      [4, 88, -100]],

      [[56, 67, -3],
      [4, 88, -100]],

      [[56, 67, -3],
      [4, 88, -100]],

      [[56, 67, -3],
      [4, 88, -100]]
    ],

], dtype='int16')
print(four_dim)

string_tensor = tf.constant(['Hello', 'World'])
print(string_tensor)
```

```python
# Rabdom Tensor
normal_random_tf = tf.random.normal(
    shape=(3,4),
    mean=1.0,
    stddev=0.01,
    dtype=tf.dtypes.float32,
    seed=None,
    name=None
)

uniform_random_tf = tf.random.uniform(
    shape=(3,3),
    minval=5,
    maxval=10,
    dtype=tf.dtypes.float32,
    seed=None,
    name=None
)
uniform_random_tf
```

```python
t1 = tf.constant([
    [[1,2,3],
    [4,5,6]]
])
t2 = tf.constant([
    [[7,8,9],
    [10,11,12]]
])

print(f"t1 shape: {t1.shape}")
print(f"t2 shape: {t2.shape}")
print(f"concat shape: {tf.concat([t1, t2], axis=0).shape}")
print(f"stack shape: {tf.stack([t1, t2], axis=0).shape}")
```
